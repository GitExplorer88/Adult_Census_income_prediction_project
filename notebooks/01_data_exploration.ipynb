```python
def classify_column(data):
    classification={}
    for col in df.columns:
        if df[col].dtype == 'object':
            classification[col] = 'Text'
        elif df[col].dtype in ['int64', 'float64']:
            classification[col] = 'Numerical'
        elif df[col].dtype == 'datetime64[ns]':
            classification[col]= 'Datetime'
        else:
            classification[col] = 'Other'
    
    return classification


column_class= classify_column(df)

column_class
```

##### Role in Dataset

 **TARGET COLUMN**
- `salary`: Earning category of the person( <=50K or >50K)

```python
def visualization_of_categorical_columns(df: pd.DataFrame, column: str):
    if df[column].dtype == 'object':
        Temp_df = df[column].value_counts(normalize=True) * 100
        temp_df = Temp_df.reset_index()
        temp_df.columns = [column, 'Percentage']
        temp_df=temp_df.sort_values(by='Percentage', ascending=False).head(10)
        vibrant_colors = sns.color_palette("Set2", n_colors=len(temp_df))

        
        plt.figure(figsize=(8, 6))
        barplot = sns.barplot(
            x='Percentage', 
            y=column, 
            data=temp_df, 
            palette=vibrant_colors, 
            hue=column,  
            dodge=False,  
            legend=False  
        )
        
        # Annotate bars with percentages
        for index, row in temp_df.iterrows():
            barplot.text(
                row['Percentage'] + 1,  
                index,                  
                f"{row['Percentage']:.1f}%",  
                color='black',
                ha='left', 
                fontsize=10
            )
        
        plt.xlabel('Percentage')
        plt.ylabel(column)
        plt.title(f'{column} Distribution', fontsize=14)
        plt.xlim(0, 100)  
        plt.tight_layout()
        plt.show()


        
for i in df.columns:
    visualization_of_categorical_columns(df,i)
```

Decide Threshold for continous vs discrete is
Discrete: Unique Values<120
and Continuous: Unique Values>=120 

```python
def Descrete_column_Frequency_tracker(df: pd.DataFrame, column: str):
    if df[column].dtype in ['int64', 'float64']:
        if len(df[column].unique()) < 120:
            Temp_df = df[column].value_counts(normalize=True) * 100
            Temp_df = Temp_df.reset_index()
            Temp_df.columns = [column, 'Percentage']
            Temp_df = Temp_df.sort_values(by='Percentage', ascending=False).head(10)
            
            
            sns.set_palette('coolwarm')
            ax = sns.barplot(x=column, y='Percentage', data=Temp_df)
            
            
            for bar in ax.patches:
                percentage = bar.get_height()  
                ax.text(
                    x=bar.get_x() + bar.get_width() / 2,  
                    y=percentage,                         
                    s=f'{percentage:.1f}%',             
                    ha='center',                         
                    va='bottom',                         
                    fontsize=10                          
                )
                
            plt.xlabel('Percentage')
            plt.ylabel(column)
            plt.title(f'{column} Distribution', fontsize=14)

            
            plt.show()


for i in df.columns:
    Descrete_column_Frequency_tracker(df, i)


```

```python
marital_status_negative_impact_on_salary = df[df['salary'] == '<=50K'].groupby('marital-status')['salary'].count().reset_index()
marital_status_positive_impact_on_salary = df[df['salary'] == '>50K'].groupby('marital-status')['salary'].count().reset_index()


marital_status_impact = pd.merge(
    marital_status_positive_impact_on_salary.rename(columns={'salary': 'more than 50k'}),
    marital_status_negative_impact_on_salary.rename(columns={'salary': 'Less than 50k'}),
    on='marital-status',
    how='outer'
)


marital_status_impact['Total Representation'] = marital_status_impact['more than 50k'] + marital_status_impact['Less than 50k']
marital_status_impact['more than 50k in percentage'] = (marital_status_impact['more than 50k'] / marital_status_impact['Total Representation']) * 100
marital_status_impact['Less than 50k in percentage'] = (marital_status_impact['Less than 50k'] / marital_status_impact['Total Representation']) * 100

x = np.arange(len(marital_status_impact['marital-status']))  
width = 0.35  
fig, ax = plt.subplots(figsize=(10, 6))

bar1 = ax.bar(x - width/2, marital_status_impact['more than 50k in percentage'], width, label='More than 50k in %', color='orange')
bar2 = ax.bar(x + width/2, marital_status_impact['Less than 50k in percentage'], width, label='Less than 50k in %', color='blue')

for bar in bar1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():.1f}', ha='center', va='bottom', fontsize=10)

for bar in bar2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():.1f}', ha='center', va='bottom', fontsize=10)

ax.set_xlabel('Marital Status')
ax.set_ylabel('Percentage')
ax.set_title('Impact of Marital Status on Income')
ax.set_xticks(x)
ax.set_xticklabels(marital_status_impact['marital-status'])  
ax.legend()

plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```
